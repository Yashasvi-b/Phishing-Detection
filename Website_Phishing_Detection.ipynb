# Step 1: Import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

# Step 2: Load the dataset

https://www.kaggle.com/datasets/shashwatwork/phishing-dataset-for-machine-learning

```python
data=pd.read_csv('phish.csv')
```

```python
data.head()
```

```python
data.max(axis=0)  #im interested in finding values per column therefore axis = 0
```

```python
data.drop('id',axis=1,inplace=True)
#we are removing id column because its not contributing much
#and to remove it premanentaly from the memory we write inplace=True
```

```python
data.head()
```

```python
data['CLASS_LABEL'].hist()
```

# Step 4: Splitting data

```python
X=data.drop("CLASS_LABEL",axis=1)#for thsi instance only not from memory
y=data["CLASS_LABEL"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

# Step 5: Training Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
preds=rfc.predict(X_test)
```

# Step 6: Model evaluation metrics
these metrics are used for mainly binary classfication

```python
preds==y_test.values #to check the predcition for y_test values after all training
```

```python
sum(preds==y_test.values)/len(preds)  #this gives us accuracy
#total prediction and hteir sum
# divided by the length of the predictions
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
accuracy_score(y_test,preds)
```

```python
precision_score(y_test,preds)
```

```python
recall_score(y_test,preds)
```
